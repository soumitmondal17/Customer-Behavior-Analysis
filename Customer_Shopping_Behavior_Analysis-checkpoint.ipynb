### Replacing the null values in 'Review Rating' column with median review rating by category.

```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

### Checking if there are any null values after the above code

```python
df.isnull().sum()
```

### Cleaning/snake casing the column headers

```python
df.columns=df.columns.str.strip().str.lower().str.replace(' ','_')
df= df.rename(columns = {'purchase_amount_(usd)':'purchase_amount'})
```

```python
df
```

### Creating new column - Age group

```python
labels=['Young Adult','Adult','Middle aged','Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head()
```

### Creating new column - Purchase_frequency_days

```python
freq={
    'Every 3 Months': 90,
    'Annually': 365,
    'Quarterly': 90,
    'Monthly' : 30,
    'Bi-Weekly': 14,
    'Fortnightly': 14,
    'Weekly': 7
}
```

```python
df['purchase_frequency_days']=df['frequency_of_purchases'].map(freq)
```

```python
df[['frequency_of_purchases','purchase_frequency_days']].head(10)
```

### Dropping irrelevant column - promo_code_used

```python
df.drop(columns=['promo_code_used'],axis=1,inplace=True)
```
